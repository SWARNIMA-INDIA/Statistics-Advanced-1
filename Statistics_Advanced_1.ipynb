1. What is a random variable in probability theory?
  - A random variable is a variable whose value depends on the outcome of a random experiment. It assigns a numerical value to each possible outcome of an experiment. Random variables help us convert real-life uncertain situations into numbers so that probability rules can be applied.
  - For example, when tossing a coin three times, the number of heads obtained is a random variable. Since the outcome is not fixed and changes with every experiment, it is called a random variable.

2. What are the types of random variables?
  - There are two main types of random variables:
    - (i) Discrete Random Variable - It takes countable values such as whole numbers. These values can be listed one by one.
    - Examples:
       - Number of students present in a class
       - Number of defective items in a batch
       - Number of heads in coin tosses
    - (ii) Continuous Random Variable - It can take any value within a given range, including decimals.
    - Examples:
      - Height of a person
      - Weight of an object
      - Time taken to complete a task

3. Explain the difference between discrete and continuous distributions.
  - A discrete distribution deals with discrete random variables and assigns probabilities to specific values. The total probability is obtained by adding probabilities of all possible values.
  - A continuous distribution deals with continuous random variables. Instead of probabilities at exact points, probabilities are calculated over an interval using a probability density function.
  - In discrete distributions, the probability at a specific value is meaningful, whereas in continuous distributions, the probability at an exact point is always zero.

4. What is a binomial distribution, and how is it used in probability?
  - A binomial distribution is a discrete probability distribution that shows the probability of getting a certain number of successes in a fixed number of independent trials. Each trial has only two possible outcomes: success or failure.
  - Conditions of binomial distribution -
  Fixed number of trials
  Only two outcomes per trial
  Probability of success remains constant
  Trials are independent
  - Uses -
  Quality control testing
  Survey analysis
  Probability of pass/fail outcomes




5. What is the standard normal distribution, and why is it important?
  - The standard normal distribution is a special case of the normal distribution where the mean (μ) is 0 and Standard deviation (σ)  is 1. It is symmetrical and bell-shaped.
  - Importance:
    - Helps in calculating probabilities using Z-scores
    - Allows comparison between different datasets
    - Widely used in hypothesis testing and confidence intervals
    - It simplifies calculations and makes statistical analysis easier.

6. What is the Central Limit Theorem (CLT), and why is it critical in statistics?
  - The Central Limit Theorem (CLT) states that when the sample size is large, the sampling distribution of the sample mean approaches a normal distribution, regardless of the shape of the original population distribution.
  - It is critical as:
    - Makes inferential statistics possible.
    - Justifies the use of normal distribution.
    - Used in hypothesis testing and estimation.

7.  What is the significance of confidence intervals in statistical analysis?
  - A confidence interval provides a range of values to estimate a population parameter.
  - It is associated with a confidence level (such as 95% or 99%).
  - It is calculated using sample data.
  - Larger sample size gives a narrower and more accurate interval.
  - It shows the reliability and precision of an estimate.
  - It is widely used in statistical inference and decision making.

8. What is the concept of expected value in a probability distribution?
  - The expected value is the average value of a random variable over a large number of trials. It represents the long-term outcome of a random experiment.
  It is calculated by multiplying each possible value by its probability and then adding all the results.
  - Uses:
    - Decision making under uncertainty
    - Risk analysis
    - Economics and finance predictions



```python
"""9. Write a Python program to generate 1000 random numbers from
 a normal distribution with mean = 50 and standard deviation = 5.
Compute its mean and standard deviation using NumPy, and draw a histogram to visualize the distribution.
(Include your Python code and output in the code box below.)"""

'''Explanation: In this program, 1000 random numbers are generated from a normal distribution
 with mean 50 and standard deviation 5 using NumPy. The mean and standard deviation of the
 generated data are calculated, and a histogram is drawn to visualize the distribution.'''

import numpy as np
import matplotlib.pyplot as plt

data = np.random.normal(loc=50, scale=5, size=1000)

mean_value = np.mean(data)
std_value = np.std(data)

print("Mean:", mean_value)
print("Standard Deviation:", std_value)

plt.hist(data, bins=30)
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Histogram of Normal Distribution")
plt.show()
```

```python
"""10. You are working as a data analyst for a retail company.
The company has collected daily sales data for 2 years and wants you to identify the overall sales trend.
daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
235, 260, 245, 250, 225, 270, 265, 255, 250, 260]
● Explain how you would apply the Central Limit Theorem to estimate the average sales
with a 95% confidence interval.
● Write the Python code to compute the mean sales and its confidence interval."""

'''The Central Limit Theorem allows us to estimate the population mean using sample data.
Here, the mean sales and 95% confidence interval are calculated using sample mean and standard error.'''

import numpy as np
from scipy import stats

sales = [220,245,210,265,230,250,260,275,240,255,
         235,260,245,250,225,270,265,255,250,260]

mean_sales = np.mean(sales)
std_sales = np.std(sales, ddof=1)
n = len(sales)

confidence_interval = stats.norm.interval(
    0.95,
    loc=mean_sales,
    scale=std_sales / np.sqrt(n)
)

print("Mean Sales:", mean_sales)
print("95% Confidence Interval:", confidence_interval)
```
